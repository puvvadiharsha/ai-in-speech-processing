```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
y, sr = librosa.load('lab6.wav')

start = 10000
end =10500
vowel_segment = y[start:end]

# Perform FFT
fft_result = np.fft.fft(vowel_segment)

# Get amplitude spectrum
amplitude_spectrum = np.abs(fft_result)

# Plot the amplitude spectrum
plt.figure(figsize=(14, 5))
plt.plot(amplitude_spectrum)
plt.title('Amplitude Spectrum of Vowel Sound')
plt.show()
```

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
y, sr = librosa.load('lab6.wav')

start = 10000
end = 10500
consonant_segment = y[start:end]
# Perform FFT
fft_result = np.fft.fft(consonant_segment)
# Get amplitude spectrum
amplitude_spectrum = np.abs(fft_result)
# Define frequency axis
freqs = np.fft.fftfreq(len(consonant_segment), 1/sr)
plt.figure(figsize=(14, 5))
plt.plot(freqs[:len(freqs)//2], amplitude_spectrum[:len(amplitude_spectrum)//2])
plt.title('Amplitude Spectrum of Consonant Sound')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.show()
```

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
y, sr = librosa.load('lab6.wav')

start = 10000
end = 10500
silence_segment = y[start:end]
# Perform FFT
fft_result = np.fft.fft(silence_segment)
# Get amplitude spectrum
amplitude_spectrum = np.abs(fft_result)
# Define frequency axis
freqs = np.fft.fftfreq(len(silence_segment), 1/sr)
plt.figure(figsize=(14, 5))
plt.plot(freqs[:len(freqs)//2], amplitude_spectrum[:len(amplitude_spectrum)//2])
plt.title('Amplitude Spectrum of Silence or Non-Voiced Portion')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
sample_rate, signal = wavfile.read('lab6.wav')
silence_start, silence_end = 30000, 35000
non_voiced_start, non_voiced_end = 40000, 45000

# Extract silence and non-voiced segments from the signal
silence_segment = signal[silence_start:silence_end]
non_voiced_segment = signal[non_voiced_start:non_voiced_end]

def plot_fft(signal, sample_rate, title):
    fft_result = np.fft.fft(signal)
    magnitude = np.abs(fft_result)
    frequency = np.fft.fftfreq(len(signal), 1/sample_rate)
    plt.figure(figsize=(10, 4))
    plt.plot(frequency[:len(frequency)//2], magnitude[:len(magnitude)//2])
    plt.title(title)
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()

plot_fft(silence_segment, sample_rate, 'FFT for Silence')
plot_fft(non_voiced_segment, sample_rate, 'FFT for Non-Voiced Sound')
```

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
y, sr = librosa.load('lab6.wav')

# Compute the spectrogram
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)

plt.figure(figsize=(14, 5))
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.show()
```
